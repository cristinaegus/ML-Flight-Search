# Análisis PCA y Preparación para Machine Learning

En esta sección vamos a:
1. **Preparar los datos** para análisis numérico
2. **Aplicar PCA** para reducir dimensionalidad y entender la estructura de los datos
3. **Visualizar** los componentes principales
4. **Preparar** los datos para modelos de machine learning

```python
# Importar librerías adicionales para PCA y Machine Learning
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

print("Librerías importadas exitosamente!")
```

```python
# Explorar los datos y preparar características numéricas
print("Información detallada del dataset:")
print(f"Forma del dataset: {data.shape}")
print(f"\nColumnas disponibles: {list(data.columns)}")
print(f"\nTipos de datos:")
print(data.dtypes)
print(f"\nValores únicos por columna:")
for col in data.columns:
    print(f"{col}: {data[col].nunique()} valores únicos")
```

```python
# Crear características numéricas para PCA usando columnas existentes

data_numeric = pd.DataFrame()

# 1. Características basadas en longitud de texto de columnas existentes
data_numeric['name_length'] = data['name'].astype(str).str.len()
data_numeric['carrier_length'] = data['carrier'].astype(str).str.len()
data_numeric['origin_length'] = data['origin'].astype(str).str.len()
data_numeric['dest_length'] = data['dest'].astype(str).str.len()

# 2. Codificar columnas categóricas
le_carrier = LabelEncoder()
data_numeric['carrier_encoded'] = le_carrier.fit_transform(data['carrier'].astype(str))

le_origin = LabelEncoder()
data_numeric['origin_encoded'] = le_origin.fit_transform(data['origin'].astype(str))

le_dest = LabelEncoder()
data_numeric['dest_encoded'] = le_dest.fit_transform(data['dest'].astype(str))

# 3. Incluir columnas numéricas originales relevantes
num_cols = ['year', 'month', 'day', 'dep_time', 'sched_dep_time', 'dep_delay', 'arr_time',
            'sched_arr_time', 'arr_delay', 'flight', 'air_time', 'distance', 'hour', 'minute']
for col in num_cols:
    data_numeric[col] = data[col]

print("Características numéricas creadas:")
print(data_numeric.head())
print(f"\nForma de datos numéricos: {data_numeric.shape}")
print(f"\nDescripción estadística:")
print(data_numeric.describe())
```

```python
# Aplicar TF-IDF a las columnas de texto principales del dataset de vuelos
# Trabajaremos con las columnas de texto disponibles: carrier, origin, dest, name

# Combinar texto de las columnas categóricas para análisis
text_combined = (data['carrier'].fillna('') + ' ' + 
                data['origin'].fillna('') + ' ' + 
                data['dest'].fillna('') + ' ' + 
                data['name'].fillna('')).str.lower()

# Aplicar TF-IDF (limitamos a 30 características para evitar alta dimensionalidad)
tfidf = TfidfVectorizer(max_features=30, ngram_range=(1,2))
tfidf_features = tfidf.fit_transform(text_combined)

# Convertir a DataFrame
tfidf_df = pd.DataFrame(tfidf_features.toarray(), 
                       columns=[f'tfidf_{i}' for i in range(tfidf_features.shape[1])])

print(f"Características TF-IDF creadas: {tfidf_df.shape}")
print(f"Características TF-IDF más importantes:")
print(tfidf.get_feature_names_out()[:10])
```

```python
# Combinar todas las características numéricas
# Resetear índices para asegurar alineación
data_numeric_reset = data_numeric.reset_index(drop=True)
tfidf_df_reset = tfidf_df.reset_index(drop=True)

# Combinar características
features_combined = pd.concat([data_numeric_reset, tfidf_df_reset], axis=1)

# Manejar valores faltantes
features_combined = features_combined.fillna(0)

print(f"Dataset combinado para PCA: {features_combined.shape}")
print(f"Características incluidas: {list(features_combined.columns)}")

# Estandarizar los datos antes del PCA
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_combined)

print(f"\nDatos estandarizados para PCA: {features_scaled.shape}")
print("Preparación completada para análisis PCA!")
```

```python
# Aplicar PCA
# Empezaremos con un número de componentes que explique al menos 95% de la varianza
pca = PCA(n_components=0.95, random_state=42)
features_pca = pca.fit_transform(features_scaled)

print(f"PCA aplicado exitosamente!")
print(f"Número de componentes principales: {pca.n_components_}")
print(f"Forma de datos transformados: {features_pca.shape}")
print(f"Varianza explicada total: {pca.explained_variance_ratio_.sum():.4f}")

# Análisis de componentes principales
print(f"\nVarianza explicada por cada componente:")
for i, var in enumerate(pca.explained_variance_ratio_[:10]):  # Mostrar primeros 10
    print(f"PC{i+1}: {var:.4f} ({var*100:.2f}%)")

# Varianza acumulada
cumsum_var = np.cumsum(pca.explained_variance_ratio_)
print(f"\nVarianza explicada acumulada (primeros 10 componentes):")
for i, cum_var in enumerate(cumsum_var[:10]):
    print(f"PC{i+1}: {cum_var:.4f} ({cum_var*100:.2f}%)")
```

```python
# Visualizar los resultados del PCA

# 1. Gráfico de varianza explicada
plt.figure(figsize=(15, 5))

# Subplot 1: Varianza explicada por componente
plt.subplot(1, 3, 1)
plt.bar(range(1, min(21, len(pca.explained_variance_ratio_) + 1)), 
        pca.explained_variance_ratio_[:20])
plt.xlabel('Componente Principal')
plt.ylabel('Varianza Explicada')
plt.title('Varianza Explicada por Componente\n(Primeros 20 componentes)')
plt.xticks(rotation=45)

# Subplot 2: Varianza acumulada
plt.subplot(1, 3, 2)
plt.plot(range(1, len(cumsum_var) + 1), cumsum_var, 'bo-', markersize=4)
plt.axhline(y=0.95, color='r', linestyle='--', label='95% varianza')
plt.axhline(y=0.90, color='g', linestyle='--', label='90% varianza')
plt.xlabel('Número de Componentes')
plt.ylabel('Varianza Explicada Acumulada')
plt.title('Varianza Explicada Acumulada')
plt.legend()
plt.grid(True, alpha=0.3)

# Subplot 3: Scatter plot de los primeros 2 componentes principales
plt.subplot(1, 3, 3)
scatter = plt.scatter(features_pca[:, 0], features_pca[:, 1], 
                     c=data_numeric['carrier_encoded'], 
                     cmap='viridis', alpha=0.6, s=20)
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.3f} varianza)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.3f} varianza)')
plt.title('Proyección en los primeros 2 PCs\n(Coloreado por Aerolínea)')
plt.colorbar(scatter, label='Aerolínea Codificada')

plt.tight_layout()
plt.show()
```

```python
# Análisis de interpretabilidad de los componentes principales
print("=== INTERPRETABILIDAD DE LOS COMPONENTES PRINCIPALES ===")

# Analizar las características más importantes para cada componente
feature_names = list(features_combined.columns)
n_components_to_analyze = min(5, pca.n_components_)

for i in range(n_components_to_analyze):
    print(f"\n--- COMPONENTE PRINCIPAL {i+1} ---")
    print(f"Varianza explicada: {pca.explained_variance_ratio_[i]:.4f} ({pca.explained_variance_ratio_[i]*100:.2f}%)")
    
    # Obtener los pesos (loadings) para este componente
    loadings = pca.components_[i]
    
    # Encontrar las características más importantes (valores absolutos más altos)
    feature_importance = list(zip(feature_names, loadings))
    feature_importance.sort(key=lambda x: abs(x[1]), reverse=True)
    
    print("Características más influyentes:")
    for j, (feature, loading) in enumerate(feature_importance[:8]):  # Top 8
        print(f"  {j+1}. {feature}: {loading:.4f}")

# Crear un heatmap de los loadings para mejor visualización
plt.figure(figsize=(12, 8))
loadings_matrix = pca.components_[:n_components_to_analyze, :]
sns.heatmap(loadings_matrix, 
            xticklabels=feature_names, 
            yticklabels=[f'PC{i+1}' for i in range(n_components_to_analyze)],
            cmap='RdBu_r', center=0, annot=False)
plt.title('Loadings de los Componentes Principales\n(Contribución de cada característica)')
plt.xlabel('Características')
plt.ylabel('Componentes Principales')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Modelo de Machine Learning

Ahora que hemos analizado la estructura de los datos con PCA, vamos a crear un modelo de machine learning. 

**Objetivo del modelo:** Predecir la aerolínea (carrier) basado en las características de los vuelos.

Esto nos permitirá:
- Clasificar automáticamente qué aerolínea es más probable para ciertas rutas y horarios
- Entender qué características de los vuelos están más asociadas con diferentes aerolíneas
- Analizar patrones de operación de las diferentes aerolíneas

```python
# Preparar datos para el modelo de Machine Learning

# Usar los datos transformados por PCA como características
X = features_pca  # Características (componentes principales)
y = data['carrier'].fillna('unknown')  # Variable objetivo (aerolíneas)

# Verificar la distribución de las clases
print("Distribución de aerolíneas (clases):")
print(y.value_counts().head(10))
print(f"\nTotal de aerolíneas únicas: {y.nunique()}")

# Para simplificar el modelo, vamos a enfocarnos en las aerolíneas más comunes
# Esto evitará problemas de clases muy desbalanceadas
min_samples = 1000  # Mínimo número de muestras por aerolínea
class_counts = y.value_counts()
common_classes = class_counts[class_counts >= min_samples].index

# Filtrar datos para incluir solo las aerolíneas comunes
mask = y.isin(common_classes)
X_filtered = X[mask]
y_filtered = y[mask]

print(f"\nDespués del filtrado:")
print(f"Número de muestras: {len(X_filtered)}")
print(f"Número de aerolíneas: {y_filtered.nunique()}")
print(f"Aerolíneas incluidas: {list(y_filtered.unique())}")

# Dividir en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X_filtered, y_filtered, test_size=0.2, random_state=42, stratify=y_filtered
)

print(f"\nDivisión de datos:")
print(f"Entrenamiento: {X_train.shape[0]} muestras")
print(f"Prueba: {X_test.shape[0]} muestras")
```

```python
# Entrenar modelo de Random Forest
print("=== ENTRENAMIENTO DEL MODELO ===")

# Crear y entrenar el modelo
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    random_state=42,
    n_jobs=-1
)

# Entrenar el modelo
print("Entrenando Random Forest...")
rf_model.fit(X_train, y_train)

# Hacer predicciones
y_pred = rf_model.predict(X_test)

# Calcular precisión
accuracy = rf_model.score(X_test, y_test)
print(f"Precisión del modelo: {accuracy:.4f} ({accuracy*100:.2f}%)")

# Reporte de clasificación detallado
print("\n=== REPORTE DE CLASIFICACIÓN ===")
print(classification_report(y_test, y_pred, zero_division=0))
```

```python
# Análisis del modelo y visualizaciones

# Asegúrate de que rf_model esté definido
try:
    rf_model
except NameError:
    raise RuntimeError("El modelo 'rf_model' no está definido. Ejecuta primero la celda de entrenamiento del modelo (Random Forest).")

plt.figure(figsize=(15, 10))

# 1. Importancia de los componentes principales
plt.subplot(2, 2, 1)
feature_importance = rf_model.feature_importances_
pc_names = [f'PC{i+1}' for i in range(len(feature_importance))]
indices = np.argsort(feature_importance)[::-1][:15]  # Top 15

plt.barh(range(len(indices)), feature_importance[indices])
plt.yticks(range(len(indices)), [pc_names[i] for i in indices])
plt.xlabel('Importancia')
plt.title('Importancia de Componentes Principales\nen el Modelo')
plt.gca().invert_yaxis()

# 2. Matriz de confusión
plt.subplot(2, 2, 2)
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=rf_model.classes_, 
            yticklabels=rf_model.classes_)
plt.title('Matriz de Confusión')
plt.xlabel('Predicción')
plt.ylabel('Valor Real')
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# 3. Distribución de predicciones
plt.subplot(2, 2, 3)
pred_counts = pd.Series(y_pred).value_counts()
plt.pie(pred_counts.values, labels=pred_counts.index, autopct='%1.1f%%')
plt.title('Distribución de Predicciones')

# 4. Comparación Real vs Predicción
plt.subplot(2, 2, 4)
comparison_df = pd.DataFrame({
    'Real': y_test.value_counts(),
    'Predicción': pd.Series(y_pred).value_counts()
}).fillna(0)

comparison_df.plot(kind='bar', ax=plt.gca())
plt.title('Comparación: Real vs Predicción')
plt.xlabel('Sitios Web')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()

# Resumen del rendimiento
print("\n=== RESUMEN DEL MODELO ===")
print(f"Modelo: Random Forest")
print(f"Características utilizadas: {X_train.shape[1]} Componentes Principales")
print(f"Precisión: {accuracy:.4f}")
print(f"Número de clases predichas: {len(rf_model.classes_)}")
print(f"Componente principal más importante: PC{indices[0] + 1}")
```
